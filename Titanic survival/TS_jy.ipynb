```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
data=pd.read_csv("tested.csv")
data.head()
```

```python
columns_to_drop = ['Name', 'PassengerId', 'Pclass', 'SibSp', 'Parch', 'Ticket', 'Cabin','Embarked']
data.drop(columns=columns_to_drop, inplace=True)
data
```

```python
X=data.Survived
Y=data.drop(['Survived'],axis='columns')
Y
```

```python
dum=pd.get_dummies(Y.Sex)
dum.head()
```

```python
Y=pd.concat([Y,dum],axis='columns')
Y.drop(['Sex'],axis='columns',inplace=True)
Y.head()
```

```python
Y.Age=Y.Age.fillna(Y.Age.mean())
Y.Fare=Y.Fare.fillna(Y.Fare.mean())
Y.columns[Y.isna().any()]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Y,X, test_size=0.2, random_state=2)
X_test
```

```python
model= GaussianNB()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```
